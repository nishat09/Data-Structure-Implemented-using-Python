```python
#Stack using array

class stack:
    stack=[]
    pointer=-1
    
    def push(self,value):
        self.stack.append(value)
        self.pointer+=1
    
    def pop(self):
        if len(self.stack)==0:
            raise Exception ('Stack is Empty')
        temp_val=self.stack[self.pointer]
        self.stack=self.stack[:-1]
        self.pointer-=1
        return temp_val
    def peek(self):
        if len(self.stack)==0:
            return "Stack is Empty"
        return self.stack[self.pointer]
    
    def is_empty(self):
        if len(self.stack)==0:
            return True
        return False

def matched(a,b):
    
    if a=='(' and b==')':
        return True
    elif a=='{' and b=='}':
        return True
    elif a=='[' and b==']':
        return True
    else:
        return False
    
    

    
def paran_checker(string):
    s=stack()
    is_balanced=True
    list1=['(','{','[']
    list2=[')','}',']']
    list_idx=[]
    cnt=0
    for i in string:
        if i in list1:
            s.push(i)
            list_idx.append(cnt)
            
        else:
            if i in list2:
                if  s.is_empty():
                    is_balanced=False
                    print('This expression is NOT correct.')
                    print('Error at character # '+str(cnt+1)+". '"+i+"'- not opened.")
                    return 

                else:
                    temp=s.peek()
                    if not matched(temp,i):
                        is_balanced=False
                    if matched(temp,i):
                        is_balanced=True
                        s.pop()
                        list_idx.pop()
        
        cnt+=1
        
    if s.is_empty and is_balanced:
        print('This expression is correct.')
        return 
    else:
        test=s.peek()
        print('This expression is NOT correct.')
        print('Error at character # '+str(list_idx[-1]+1)+". '"+test+"'- not closed.")
        return

string=input()
paran_checker(string)
                
        
        
    
```

```python
#Stack using linked list
class node:
    def __init__(self,value):
        self.value=value
        self.nextValue=None

        
class stack:
    head=None
    
    def push(self,value):
        if self.head==None:
            new_node=node(value)
            self.head=new_node
        else:
            new_node=node(value)
            new_node.nextValue=self.head
            self.head=new_node
    def pop(self):
        if self.head==None:
            raise Exception ("Stack is empty")
        temp=self.head
        value=temp.value
        self.head=self.head.nextValue
        temp=None
        return value
    def peek(self):
        return self.head.value
    
    def is_empty(self):
        if self.head ==None:
            return True
        return False
    

def matched(a,b):
    
    if a=='(' and b==')':
        return True
    elif a=='{' and b=='}':
        return True
    elif a=='[' and b==']':
        return True
    else:
        return False
    
    

    
def paran_checker(string):
    s=stack()
    is_balanced=True
    list1=['(','{','[']
    list2=[')','}',']']
    list_idx=[]
    cnt=0
    for i in string:
        if i in list1:
            s.push(i)
            list_idx.append(cnt)
            
        else:
            if i in list2:
                if  s.is_empty():
                    is_balanced=False
                    print('This expression is NOT correct.')
                    print('Error at character # '+str(cnt+1)+". '"+i+"'- not opened.")
                    return 

                else:
                    temp=s.peek()
                    if not matched(temp,i):
                        is_balanced=False
                    if matched(temp,i):
                        is_balanced=True
                        s.pop()
                        list_idx.pop()
        
        cnt+=1
        
    if s.is_empty and is_balanced:
        print('This expression is correct.')
        return 
    else:
        test=s.peek()
        print('This expression is NOT correct.')
        print('Error at character # '+str(list_idx[-1]+1)+". '"+test+"'- not closed.")
        return

string=input()
paran_checker(string)

        
        
```
